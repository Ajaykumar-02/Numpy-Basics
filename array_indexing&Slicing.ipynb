Numpy Indexing & Slicing " Vectors "

```python
import numpy as np
```

```python
arr=np.arange(11,21)
arr
```

```python
arr[5] #indexing
```

```python
arr[1:5] #slicing
```

```python
arr[3::2]
```

Numpy Indexing & Slicing " Matrix "

```python
arr=np.arange(1,31).reshape(6,5)
arr
```

```python
arr[3,2] #3=row & 2=column/index
```

arr[2,3]

```python
slice=arr[2:4,2:4]
```

```python
slice
```

```python
arr[3:6,3:5]
```

```python
arr[:,4]
```

```python
arr[3:,3:]
```

Boolean indexing

```python
arr=np.arange(11,21)
arr
```

```python
bool_index=arr%2==0 #condition of even no.
bool_index #o/p= True or False
```

```python
arr=arr[bool_index] #arr[index=bool_index]
arr #even no.
```
